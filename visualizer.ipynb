```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
x_file_name = 'laser_power_grid_search_medium_power.txt'
y_file_name = 'relaxation_grid_search_medium_power.txt'
z_file_name = 'magnitude_grid_search_medium_power.txt'
SNR_file_name = 'SNR_grid_search_medium_power.txt'

x = np.array(np.loadtxt(x_file_name))
y = np.array(np.loadtxt(y_file_name))

z = np.loadtxt(z_file_name)
z = np.array(z).reshape(len(x), len(y))

SNR = np.loadtxt(SNR_file_name)
SNR = np.array(SNR).reshape(len(x), len(y))

# Prepare DataFrames
df_magnitude = pd.DataFrame(z, index=y, columns=x)
df_SNR = pd.DataFrame(SNR, index=y, columns=x)

# Create subplots
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 10))

# Plot magnitude heatmap
sns.heatmap(df_magnitude, ax=axes[0], cmap="inferno", cbar_kws={'label': 'Magnitude'})
axes[0].set_xlabel("Laser Power")
axes[0].set_ylabel("Relaxation Rate")
axes[0].set_title("Heatmap of Magnitude")
axes[0].tick_params(axis='x', rotation=45)

# Plot SNR heatmap
sns.heatmap(df_SNR, ax=axes[1], cmap="viridis", cbar_kws={'label': 'SNR'})
axes[1].set_xlabel("Laser Power")
axes[1].set_ylabel("Relaxation Rate")
axes[1].set_title("Heatmap of SNR")
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
