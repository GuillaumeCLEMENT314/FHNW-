# Quantum Bloch equation

## Definition of the physical system

We are here solving the quantum Bloch equation for $^{87}\text{Rb}$  in order to get a more precise estimation of our needs in pump field (optical and RF) in order to successfully build the self-oscillating OPM. 
The expression we will be using comes directly from the paper "Rubidium 87 D Line Data" by Daniel A. Steck.

Here is the general expression:

\begin{align*}
\frac{\partial}{\partial t} \tilde{\rho}_{\alpha\, m_\alpha,\, \beta\, m_\beta}
&= -\frac{i}{2} \bigg[
    \delta_{\alpha e} \sum_{m_g} \Omega(m_\alpha, m_g)\, \tilde{\rho}_{g\, m_g,\, \beta\, m_\beta}
    - \delta_{g \beta} \sum_{m_e} \Omega(m_e, m_\beta)\, \tilde{\rho}_{\alpha\, m_\alpha,\, e\, m_e} \\
&\quad
    + \delta_{\alpha g} \sum_{m_e} \Omega^*(m_e, m_\alpha)\, \tilde{\rho}_{e\, m_e,\, \beta\, m_\beta}
    - \delta_{e \beta} \sum_{m_g} \Omega^*(m_\beta, m_g)\, \tilde{\rho}_{\alpha\, m_\alpha,\, g\, m_g}
\bigg] \\
&\quad
    - \delta_{\alpha e} \delta_{e \beta} \, \Gamma\, \tilde{\rho}_{\alpha\, m_\alpha,\, \beta\, m_\beta} \\
&\quad
    - \delta_{\alpha e} \delta_{g \beta} \, \frac{\Gamma}{2}\, \tilde{\rho}_{\alpha\, m_\alpha,\, \beta\, m_\beta} \\
&\quad
    - \delta_{\alpha g} \delta_{e \beta} \, \frac{\Gamma}{2}\, \tilde{\rho}_{\alpha\, m_\alpha,\, \beta\, m_\beta} \\
&\quad
    + \delta_{\alpha g} \delta_{g \beta} \, \Gamma \sum_{q=-1}^{1}
    \tilde{\rho}_{e\,(m_\alpha+q),\, e\,(m_\beta+q)}\,
    \langle F_e\, (m_\alpha+q) | F_g\, 1\, m_\alpha\, q \rangle\,
    \langle F_e\, (m_\beta+q) | F_g\, 1\, m_\beta\, q \rangle \\
&\quad
    + i\, (\delta_{\alpha e} \delta_{g \beta} - \delta_{\alpha g} \delta_{e \beta})\, \Delta\,
    \tilde{\rho}_{\alpha\, m_\alpha,\, \beta\, m_\beta}
\end{align*}

where
\begin{align*}
\Omega(m_e, m_g) 
&= \langle F_g \, m_g | F_e \, 1 \, m_e - (m_e - m_g) \rangle\, \Omega_{-(m_e - m_g)} \\
&= (-1)^{F_e - F_g + m_e - m_g} \sqrt{\frac{2F_g + 1}{2F_e + 1}}\, \langle F_e \, m_e | F_g \, 1 \, m_g \, (m_e - m_g) \rangle\, \Omega_{-(m_e - m_g)}
\end{align*}

is the Rabi frequency between two magnetic sublevels, and
\begin{align*}
\Omega_q = \frac{2 \langle F_e || er || F_g \rangle E_q^{(+)}}{\hbar}
\end{align*}

It is important to precise that the indices  $\alpha, m_{\alpha}, \beta, m_{\beta}$ represent respectively our initial state and our final state for the transition of interest.

We can break down this equation into three categories. <br>
**Pumping** :
\begin{align*}
P&= -\frac{i}{2} \bigg[
    \delta_{\alpha e} \sum_{m_g} \Omega(m_\alpha, m_g)\, \tilde{\rho}_{g\, m_g,\, \beta\, m_\beta}
    - \delta_{g \beta} \sum_{m_e} \Omega(m_e, m_\beta)\, \tilde{\rho}_{\alpha\, m_\alpha,\, e\, m_e} \\
&\quad
    + \delta_{\alpha g} \sum_{m_e} \Omega^*(m_e, m_\alpha)\, \tilde{\rho}_{e\, m_e,\, \beta\, m_\beta}
    - \delta_{e \beta} \sum_{m_g} \Omega^*(m_\beta, m_g)\, \tilde{\rho}_{\alpha\, m_\alpha,\, g\, m_g}
\bigg] \quad 
\\
\end{align*}

**Dissipation** :
\begin{align*}
D = & - \delta_{\alpha e} \delta_{e \beta} \, \Gamma\, \tilde{\rho}_{\alpha\, m_\alpha,\, \beta\, m_\beta} \\
\quad
    &- \delta_{\alpha e} \delta_{g \beta} \, \frac{\Gamma}{2}\, \tilde{\rho}_{\alpha\, m_\alpha,\, \beta\, m_\beta} \\
\quad
    &- \delta_{\alpha g} \delta_{e \beta} \, \frac{\Gamma}{2}\, \tilde{\rho}_{\alpha\, m_\alpha,\, \beta\, m_\beta} \\
\quad
    &+ \delta_{\alpha g} \delta_{g \beta} \, \Gamma \sum_{q=-1}^{1}
    \tilde{\rho}_{e\,(m_\alpha+q),\, e\,(m_\beta+q)}\,
    \langle F_e\, (m_\alpha+q) | F_g\, 1\, m_\alpha\, q \rangle\,
    \langle F_e\, (m_\beta+q) | F_g\, 1\, m_\beta\, q \rangle 
&\quad 
\end{align*}

**Free Evolution** :
\begin{align*}
E = &\quad + i\, (\delta_{\alpha e} \delta_{g \beta} - \delta_{\alpha g} \delta_{e \beta})\, \Delta\,
    \tilde{\rho}_{\alpha\, m_\alpha,\, \beta\, m_\beta}
    \quad
\end{align*}

We are now interested in multiple transitions. For each of them, we have to run the equatio (that will get largely simplified depending on the type of transition). We will describe separately the transitions for the magnetic sublevels transitions and the optical ones.

### Magnetic sublevels

We will start by making the equations for every transition in our magnetic sublevels at the ground state. We want to be able to jump from one to the upper or lower one. We thus have 8 possible transitions:
\begin{alignat}{3}
    &\ket{\alpha, \, m_{\alpha}} \quad &\rightarrow\quad & \ket{\beta, \, m_{\beta}} \nonumber \\
    &\ket{F, \, m_F = -2} \quad &\rightarrow\quad & \ket{F, \, m_F = -1} \\
    &\ket{F, \, m_F = -1} \quad &\rightarrow\quad & \ket{F, \, m_F = 0}  \\
    &\ket{F, \, m_F = 0}  \quad &\rightarrow\quad & \ket{F, \, m_F = +1} \\
    &\ket{F, \, m_F = +1} \quad &\rightarrow\quad & \ket{F, \, m_F = +2} \\
    &\ket{F, \, m_F = +2} \quad &\rightarrow\quad & \ket{F, \, m_F = +1} \\
    &\ket{F, \, m_F = +1} \quad &\rightarrow\quad & \ket{F, \, m_F = 0}  \\
    &\ket{F, \, m_F = 0}  \quad &\rightarrow\quad & \ket{F, \, m_F = -1} \\
    &\ket{F, \, m_F = -1} \quad &\rightarrow\quad & \ket{F, \, m_F = -2}
\end{alignat}


Now we have the tedious task to compute the quantum bloch equation for every transition of interest. For the sake of simplicity, only the final result will be written here. The general form can be given, since the kronecker function is about ground and excited states, and here we only have ground states.

General form:

\begin{align*}
    \partial_t \rho_{\alpha \, m_{\alpha}, \beta \, m_{\beta}} = & - \frac{i}{2} \left[-\sum_{m_e} \Omega(m_e,m_{\beta}) \rho_{\alpha \, m_{\alpha}, e \, m_e} + \sum_{m_e} \Omega^*(m_e, m_{\alpha}) \rho_{e \, m_e, \beta \, m_{\beta}} \right] \\
    &+ \Gamma \sum^1_{q=-1} \rho_{e(m_{\alpha}+q),e(m_{\beta}+q)} \langle{F' \, (m_{\alpha}+q)}|{F \, 1 \, m_{\alpha} \, q}\rangle \langle{F'\, (m_{\beta}+q)}| {F \, 1 \, m_{\beta} \, q}\rangle
\end{align*}

(1):

\begin{align*}
    \partial_t \rho_{g \, -2, g \, -1} = & - \frac{i}{2} \left[-\sum_{m_e} \Omega(m_e,-1) \rho_{g \, -2, e \, m_e} + \sum_{m_e} \Omega^*(m_e, -2) \rho_{e \, m_e, g \, -1} \right] \\
    &+ \Gamma \sum^1_{q=-1} \rho_{e(-2+q),e(-1+q)} \langle{F' \, (-2+q)}|{F \, 1 \, -2 \, q}\rangle \langle{F'\, (-1+q)}| {F \, 1 \, -1 \, q}\rangle
\end{align*}

(2):

\begin{align*}
    \partial_t \rho_{g \, -1, g \, 0} = & - \frac{i}{2} \left[-\sum_{m_e} \Omega(m_e,0) \rho_{g \, -1, e \, m_e} + \sum_{m_e} \Omega^*(m_e, -1) \rho_{e \, m_e, g \, 0} \right] \\
    &+ \Gamma \sum^1_{q=-1} \rho_{e(-1+q),e(0+q)} \langle{F' \, (-1+q)}|{F \, 1 \, -1 \, q}\rangle \langle{F'\, (0+q)}| {F \, 1 \, 0 \, q}\rangle
\end{align*}

(3):

\begin{align*}
    \partial_t \rho_{g \, 0, g \, +1} = & - \frac{i}{2} \left[-\sum_{m_e} \Omega(m_e,+1) \rho_{g \, 0, e \, m_e} + \sum_{m_e} \Omega^*(m_e, 0) \rho_{e \, m_e, g \, +1} \right] \\
    &+ \Gamma \sum^1_{q=-1} \rho_{e(0+q),e(+1+q)} \langle{F' \, (0+q)}|{F \, 1 \, 0 \, q}\rangle \langle{F'\, (+1+q)}| {F \, 1 \, +1 \, q}\rangle
\end{align*}

(4):

\begin{align*}
    \partial_t \rho_{g \, +1, g \, +2} = & - \frac{i}{2} \left[-\sum_{m_e} \Omega(m_e,+2) \rho_{g \, +1, e \, m_e} + \sum_{m_e} \Omega^*(m_e, +1) \rho_{e \, m_e, g \, +2} \right] \\
    &+ \Gamma \sum^1_{q=-1} \rho_{e(+1+q),e(+2+q)} \langle{F' \, (+1+q)}|{F \, 1 \, +1 \, q}\rangle \langle{F'\, (+2+q)}| {F \, 1 \, +2 \, q}\rangle
\end{align*}

(5):

\begin{align*}
    \partial_t \rho_{g \, +2, g \, +1} = & - \frac{i}{2} \left[-\sum_{m_e} \Omega(m_e,+1) \rho_{g \, +2, e \, m_e} + \sum_{m_e} \Omega^*(m_e, +2) \rho_{e \, m_e, g \, +1} \right] \\
    &+ \Gamma \sum^1_{q=-1} \rho_{e(+2+q),e(+1+q)} \langle{F' \, (+2+q)}|{F \, 1 \, +2 \, q}\rangle \langle{F'\, (+1+q)}| {F \, 1 \, +1 \, q}\rangle
\end{align*}

(6):

\begin{align*}
    \partial_t \rho_{g \, +1, g \, 0} = & - \frac{i}{2} \left[-\sum_{m_e} \Omega(m_e,0) \rho_{g \, +1, e \, m_e} + \sum_{m_e} \Omega^*(m_e, +1) \rho_{e \, m_e, g \, 0} \right] \\
    &+ \Gamma \sum^1_{q=-1} \rho_{e(+1+q),e(0+q)} \langle{F' \, (+1+q)}|{F \, 1 \, +1 \, q}\rangle \langle{F'\, (0+q)}| {F \, 1 \, 0 \, q}\rangle
\end{align*}

(7):

\begin{align*}
    \partial_t \rho_{g \, 0, g \, -1} = & - \frac{i}{2} \left[-\sum_{m_e} \Omega(m_e,-1) \rho_{g \, 0, e \, m_e} + \sum_{m_e} \Omega^*(m_e, 0) \rho_{e \, m_e, g \, -1} \right] \\
    &+ \Gamma \sum^1_{q=-1} \rho_{e(0+q),e(-1+q)} \langle{F' \, (0+q)}|{F \, 1 \, 0 \, q}\rangle \langle{F'\, (-1+q)}| {F \, 1 \, -1 \, q}\rangle
\end{align*}

(8):

\begin{align*}
    \partial_t \rho_{g \, -1, g \, -2} = & - \frac{i}{2} \left[-\sum_{m_e} \Omega(m_e,-2) \rho_{g \, -1, e \, m_e} + \sum_{m_e} \Omega^*(m_e, -1) \rho_{e \, m_e, g \, -2} \right] \\
    &+ \Gamma \sum^1_{q=-1} \rho_{e(-1+q),e(-2+q)} \langle{F' \, (-1+q)}|{F \, 1 \, -1 \, q}\rangle \langle{F'\, (-2+q)}| {F \, 1 \, -2 \, q}\rangle
\end{align*}

### Orbital excitations

We now take care of the optical pumping from a given magnetic sublevel to an excited one. Since we use positively circularly polarized light $\sigma^+$, we will only allow the transition from $m_F=x \rightarrow m_{F'}=x+1$.
