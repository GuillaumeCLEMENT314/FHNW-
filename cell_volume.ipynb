```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.constants import *
```

```python
P = np.linspace(2e3,3e5,1000) #Pa
T = 273.15+60 #K
# number density [cm^-3]
n_particles_cm3 = 1 / T * 10**(21.866 + 4.312 - 4040 / T)

# Convert to [particles/m^3]
n_particles_m3 = n_particles_cm3 * 1e6

# Convert to moles/m^3
n_mol_per_m3 = n_particles_m3 / N_A

V = n_mol_per_m3 * R * T / P

V_cell_CSEM = 3e-3*3e-3*1.9e-3
```

```python
index_closest = np.argmin(np.abs(V - V_cell_CSEM))
P_match = P[index_closest]
V_match = V[index_closest]
%matplotlib widget
# Plotting
plt.figure(figsize=(8, 5))
plt.plot(P, V, label='V(P)', color='blue')
plt.axhline(y=V_cell_CSEM, color='red', linestyle='--', label='V_cell_CSEM')
plt.plot(P_match, V_match, 'ko', label='Intersection Point')
plt.text(P_match, V_match * 1.1,
         f"P  {P_match/1000:.1f} kPa\nV  {V_match:.2e} m",
         ha='center', va='bottom')

plt.xlabel('Pressure (Pa)')
plt.ylabel('Volume (m)')
plt.title('Volume vs Pressure')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
